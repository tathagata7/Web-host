```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import seaborn as sns
```

```python
iris=pd.read_csv("iris_data.csv",names=['sepal length','sepal width','petal length','petal width','class'])
iris
```

```python
plt.hist(iris['sepal length'],bins=60,color='r')
```

```python
sns.distplot(iris['sepal length'],bins=60,color='g')
```

```python
#Code to plot histogram
sns.FacetGrid(iris,hue="class",height=4).map(sns.distplot,"sepal length").add_legend();
plt.show()
sns.FacetGrid(iris,hue="class",height=4).map(sns.distplot,"petal length").add_legend();
plt.show()
sns.FacetGrid(iris,hue="class",height=4).map(sns.distplot,"sepal width").add_legend();
plt.show()
sns.FacetGrid(iris,hue="class",height=4).map(sns.distplot,"petal width").add_legend();
plt.show();
```
