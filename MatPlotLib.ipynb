```python
import numpy as np
import matplotlib.pyplot as plt
a=np.arange(0,5)
b = a**2
plt.plot(a,b,'bo')
plt.ylabel("Square")
plt.xlabel("Number")
plt.grid()
plt.show()
```

```python
x = np.arange(0,5,0.2)
x
```

```python
import matplotlib.pyplot as plt
plt.plot(x,x**2,'b--',label='^2')
plt.plot(x,x**3,'rs',label='^3')
plt.plot(x,x**4,'go',label='^4')
plt.plot(x,x**5,linewidth='2',label='^2')
plt.grid()
plt.legend()
plt.show()
```

```python
x1=np.arange(1,5)
y1=[1,4,9,16]
x2=np.arange(1,5)
y2=[2,4,6,8]
```

```python
lines = plt.plot(x1,y1,x2,y2)
plt.setp(lines[0],color='r',linewidth='2')
plt.setp(lines[1],color='b',linewidth='2')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(1)
plt.subplot(121)
plt.plot([2,4,6],color='c')
plt.subplot(122)
plt.plot([1,2,3],color='m')

plt.figure(2)
x=np.arange(0,2,0.2)
y=np.arange(0,3,0.2)
plt.subplot(121)
plt.plot(x,color='r')
plt.subplot(122)
plt.plot(y,color='g')
```
